```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
import pickle
import streamlit as st
import requests
import random

def fetch_poster(movie_id):
    url = "https://api.themoviedb.org/3/movie/{}?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US".format(movie_id)
    data = requests.get(url)
    data = data.json()
    poster_path = data['poster_path']
    full_path = "https://image.tmdb.org/t/p/w500/" + poster_path
    return full_path

def recommend(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    recommended_movie_names = []
    recommended_movie_posters = []
    for i in distances[1:6]:
        # fetch the movie poster
        movie_id = movies.iloc[i[0]].movie_id
        recommended_movie_posters.append(fetch_poster(movie_id))
        recommended_movie_names.append(movies.iloc[i[0]].title)

    return recommended_movie_names,recommended_movie_posters


st.header('Movie Recommender System')
movies = pickle.load(open('movie_list.pkl','rb'))
similarity = pickle.load(open('similarity.pkl','rb'))

movie_list = movies['title'].values
selected_movie = st.selectbox(
    "Type or select a movie from the dropdown",
    movie_list
)

if st.button('Show Recommendation'):
    recommended_movie_names,recommended_movie_posters = recommend(selected_movie)
    col1, col2, col3, col4, col5 = st.columns(5)
    with col1:
        st.text(recommended_movie_names[0])
        st.image(recommended_movie_posters[0])
    with col2:
        st.text(recommended_movie_names[1])
        st.image(recommended_movie_posters[1])

    with col3:
        st.text(recommended_movie_names[2])
        st.image(recommended_movie_posters[2])
    with col4:
        st.text(recommended_movie_names[3])
        st.image(recommended_movie_posters[3])
    with col5:
        st.text(recommended_movie_names[4])
        st.image(recommended_movie_posters[4])


def get_genres():
    genre_url = f"https://api.themoviedb.org/3/genre/movie/list?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US"
    response = requests.get(genre_url).json()
    return {genre["id"]: genre["name"] for genre in response.get("genres", [])}  # Returns {id: name} dictionary


# Function to search movies using TMDb API with filters
def search_movies(query, genre_id=None, year=None):
    search_url = f"https://api.themoviedb.org/3/search/movie?api_key=8265bd1679663a7ea12ac168da84d2e8&query={query}"
    response = requests.get(search_url).json()

    # Filter movies based on genre and year
    filtered_movies = [
        movie for movie in response.get("results", [])
        if (genre_id is None or genre_id in movie.get("genre_ids", [])) and
           (year is None or movie.get("release_date", "")[:4] == str(year))
    ]

    return filtered_movies


# Function to fetch a random popular movie
def get_random_movie():
    popular_url = f"https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={random.randint(1, 10)}"
    response = requests.get(popular_url).json()
    return random.choice(response.get("results", [])) if response.get("results") else None


genres_dict = get_genres()


# Sidebar Filters
st.sidebar.header("🎛 Filters")
selected_genre = st.sidebar.selectbox("🎭 Select Genre", ["Any"] + list(genres_dict.values()))
selected_year = st.sidebar.slider("📅 Select Year", 1980, 2025, step=1, value=2020)

# Convert selected genre name to genre ID
genre_id = next((g_id for g_id, g_name in genres_dict.items() if g_name == selected_genre),
                None) if selected_genre != "Any" else None

# "Surprise Me" Button
if st.button("🎲 Surprise Me!"):
    random_movie = get_random_movie()
    if random_movie:
        st.subheader("🎉 Surprise Movie Recommendation:")
        st.image(f"https://image.tmdb.org/t/p/w300{random_movie.get('poster_path', '')}", width=200)
        st.write(f"**{random_movie['title']} ({random_movie.get('release_date', '')[:4]})**")
        st.write(f"⭐ {random_movie.get('vote_average', 0)}/10")
        st.write(f"📖 Overview: {random_movie.get('overview', 'No description available.')}")
    else:
        st.write("❌ Could not fetch a random movie. Try again.")

# Show Search Results based on filters
movie_query= []
if movie_query:
    movies = search_movies(movie_query, genre_id, selected_year)

    if movies:
        st.subheader(f"🎥 Search Results for '{movie_query}' with Filters:")
        cols = st.columns(3)
        for i, movie in enumerate(movies[:6]):
            with cols[i % 3]:
                st.image(f"https://image.tmdb.org/t/p/w200{movie.get('poster_path', '')}", width=150)
                st.write(f"**{movie['title']} ({movie.get('release_date', '')[:4]})**")
                st.write(f"⭐ {movie.get('vote_average', 0)}/10")
                st.write("---")
    else:
        st.write("❌ No movies found with these filters. Try a different search.")
```
